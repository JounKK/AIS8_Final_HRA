## 시간 데이터 전처리

```python
# null 값 0 으로 채우기
# inplace=True 사용하면 df=df.fillna(0)와 동일한 기능임
in_time.fillna(0, inplace=True)
out_time.fillna(0, inplace= True)
```

```python
in_time.head()
```

```python
# datetime 형식으로 바꾸기(직원ID행 제외) >> 0 값은 1970-01-01로 처리됨
# errors = coerce : 만약 숫자로 변경할 수 없는 데이터라면 기존 데이터를 지우고 NaN으로 설정하여 반환합니다

in_time.iloc[:, 1:] = in_time.iloc[:, 1:].apply(pd.to_datetime, errors='coerce')
out_time.iloc[:, 1:] = out_time.iloc[:, 1:].apply(pd.to_datetime, errors='coerce')
```

```python
in_time.head()
```

```python
# 행으로 in_time, out_time join
total_time = in_time.append(out_time)
```

```python
# total_time 8820 행 되었음
total_time
```

```python
total_time.max()
```

```python
# out-in time 차이 구하기
# pandas의 diff(periods=1,axis=0)사용
total_time = total_time.diff(periods=4410)
```

```python
total_time = total_time.iloc[4410:]
```

```python
total_time
```

```python
total_time.reset_index(inplace=True)
total_time.head()
```

```python
# 값이 0인 열이 몇 개 있음. 해당 날짜에 사무실이 휴무였음.
# 평균 시간을 계산 위해서 id열 제거(인덱스는 직원 ID이므로 나중에 다시 추가해야 함)

total_time.drop(columns=['index','Unnamed: 0'],axis=1,inplace=True)
total_time.head()
```
